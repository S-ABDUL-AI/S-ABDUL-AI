<a id=" 6.3 Removing Outliers"></a>
# <left> <p style="background-color:white; font-family:Hack; font-size:60%; color:brown; text-align:center; border-radius: 0px; padding:0px;width:20%"> 6.3 Removing Outliers</p> </left>

```python
# features_with_outlier: we can learn this from the above chart
features_with_outliers = ['sqft_living', 'sqft_above', 'sqft_lot', 'bathrooms', 'bedrooms']
```

```python
# Defining a function to handle outliers using the IQR method
def handle_outliers_with_iqr(df, feature):
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Using .loc to avoid SettingWithCopyWarning
    df.loc[(df[feature] < lower_bound) | (df[feature] > upper_bound), feature] = df[feature].median()

# Applying the outlier handling function to the selected features
features_with_outliers = ['sqft_living', 'sqft_above', 'sqft_lot', 'bathrooms', 'bedrooms']
for feature in features_with_outliers:
    handle_outliers_with_iqr(df, feature)
```

<a id=" 7.2 Feature  Selection"></a>
# <left> <p style="background-color:white; font-family:Hack; font-size:60%; color:brown; text-align:center; border-radius: 0px; padding:0px;width:18%">7.2 Feature Selection</p> </left>

```python
from sklearn.feature_selection import mutual_info_regression

def calculate_mi_scores(df, Logprice):
   
    # Encoding categorical variables (if present)
    for colname in X.select_dtypes(['object', 'category']):
        X[colname], _ = X[colname].factorize()
    
    # Identifying discrete features
    discrete_features = [pd.api.types.is_integer_dtype(t) for t in X.dtypes]
    
    # Calculating Mutual Information scores
    mi_scores = mutual_info_regression(X, y, discrete_features=discrete_features, random_state=0)
    
    # Creating a Pandas Series with feature names as the index
    mi_scores = pd.Series(mi_scores, name='Mutual Information Score', index=X.columns)
    
    # Sorting features by MI score in descending order
    mi_scores = mi_scores.sort_values(ascending=False)
    
    return mi_scores

mi_scores = calculate_mi_scores(df, 'Logprice')
styled_mi_scores = mi_scores.reset_index().rename(columns={'index': 'Feature'}).style\
    .set_properties(**{'background-color': '#7E354D', 'color': 'white'})

styled_mi_scores
```
