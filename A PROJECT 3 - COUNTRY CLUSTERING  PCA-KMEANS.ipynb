<h3 style='padding:10px; background:white;color:black;border-radius:0px; text-align:center;'>
     EXECUTIVE SUMMARY
</h3>

<blockquote style="background-color: black; padding: 20px; border-radius: 5px; color: white;">
    <h4>Background</h4>
    <p>The global socio-economic landscape is diverse and complex, with countries exhibiting varying patterns and disparities. In the field of socio-economic analyses and development studies, clustering emerges as a crucial method for understanding and categorizing countries based on their socio-economic indicators. </p>
    <h4>Problem Statement</h4>
    <p>The challenge lies in comprehending the intricate relationships and disparities among countries, making it difficult to formulate effective policies, allocate resources efficiently, and address specific socio-economic challenges. </p>
    <h4>Main Objective</h4>
    <p>The main objective is to employ clustering techniques to categorize countries into distinct groups based on socio-economic indicators, in order to inform policy. </p>
    <h4>Specific Objectives</h4>
    <ol>
        <li>To identify common patterns and similarities among countries using clustering methods</li>
        <li>To utilize clustering results to tailor targeted policies that address the specific needs of countries within similar clusters</li>
        <li>To simplify the complex global socio-economic landscape by grouping countries with similar cluster characteristics</li>
        <li>To improve the efficient allocation of resources by considering clusters with shared challenges and opportunities</li>
    </ol>
    <h4>Methodology</h4>
    <p><strong>Data Collection:</strong> Secondary data was collected from Kaggle Datasets</p>
    <p><strong>Data Preprocessing:</strong></p>
    <ul>
        <li>Handled missing values and removed duplicate rows</li>
    </ul>
    <p><strong>Model Selection:</strong> The Principal Component Analysis was computed utilizing the following clustering techniques:</p>
    <ul>
        <li>Kmeans</li>
        <li>DBSCAN</li>
        <li>MiniBatchKmeans</li>
        <li>Mean Shift</li>
    </ul>
    <p><strong>Model Results:</strong> The following clusters recorded the best results:</p>
    <ul>
        <li>KMeans with (n_clusters = 2 and n_clusters = 3) </li>
        <li>MiniBatchKmeans with (n_clusters = 2) </li>
    </ul>
    <ul>
  <li><strong>Kmeans(n_clusters = 3) proved Best:</strong> Thus was used to classify the countries as <strong>Low-income</strong>, <strong>Middle-income</strong> and <strong>High-income.</strong></li>
    </ul>
    <h4>Cluster Analysis Results Summary</h4>
    <h5>Low-income Countries (Cluster 0) :</h5>
    <ul>
        <li><strong>Characteristics:</strong> High child mortality rates, low income and GDP per capita, limited health expenditure, higher inflation and total fertility rates. </li>
        <li><strong>Recommendations:</strong> Policy interventions should be targeted at improving healthcare infrastructure, implementing economic policies to boost income and GDP, family planning and education programs. </li>
    </ul>
    <h5>Middle-income Countries (Cluster 1) :</h5>
    <ul>
        <li><strong>Characteristics:</strong> Moderate child mortality rates, varied income levels, diverse health expenditure, moderate inflation and total fertility rates. </li>
        <li><strong>Recommendations:</strong> Policy interventions should be geared towards strengthening healthcare and diversifying economic systems to enhance growth and sustainability. </li>
    </ul>
    <h5>High-income Countries (Cluster 2) :</h5>
    <ul>
        <li><strong>Characteristics:</strong> Low child mortality rates, high income and GDP per capita, significant health expenditure, lower inflation and total fertility rates. </li>
        <li><strong>Recommendations:</strong> Countries within this cluster should strive at maintaining high-quality healthcare and exploring opportunities for innovation and technological advancement for sustained economic prosperity. </li>
    </ul>
    <h4>Overall Recommendations</h4>
    <ul>
        <li><strong>International Collaboration:</strong> Policy interventions should encourage collaboration between countries to share best practices and facilitate knowledge transfer. </li>
        <li><strong>Investment in Healthcare:</strong> Interventions should prioritize investments in healthcare infrastructure across income categories, support health programs to mitigate child mortality and improve life expectancy. </li>
        <li><strong>Economic Diversification:</strong> Middle-income countries should focus on economic diversification for sustained growth. </li>
        <li><strong>Education and Family Planning:</strong> Population interventions should promote education, particularly in low-income countries, to break the cycle of poverty, family planning initiatives to manage population growth sustainably. </li>
     <li><strong>Model Refinement:</strong> We should ensure continuous refinement of clustering methodologies to improve model performance.</li>

    <h3>Data Description</h3>
    <ul>
    <li><strong>country:</strong> Name of the country</li>
    <li><strong>child_mort:</strong> Death of children under 5 years of age per 1000 live births</li>
     <li><strong>exports:</strong> Exports of goods and services per capita. Given as %age of the GDP per capita</li>
     <li><strong>health:</strong> Total health spending per capita. Given as %age of GDP per capita</li>
    <li><strong>imports:</strong> Imports of goods and services per capita. Given as %age of the GDP per capita</li> 
    <li><strong>Income:</strong> Net income per person</li>
    <li><strong>Inflation:</strong> The measurement of the annual growth rate of the Total GDP</li>
    <li><strong>life_expec:</strong> The average number of years a new born child would live if the current mortality patterns are to remain the same</li>
    <li><strong>total_fer:</strong> The number of children that would be born to each woman if the current age-fertility rates remain the same.</li>
    <li><strong>gdpp:</strong> The GDP per capita. Calculated as the Total GDP divided by the total population.
</li>      
    </ul>

# <left> <p style="background-color:white; font-family:Hack; font-size:70%; color:black; text-align:center; border-radius: 0px; padding:0px;width:30%"> 7.1.1Inertia vs Number of Clusters</p> </left>

```python
kMs_set = {
    'init' : 'random',
    'n_init' : 10,
    'max_iter' : 300,
    'random_state' : 42
}

kMs_list = []

for k in range(2, 10) :
    kmeans = KMeans(n_clusters=k, **kMs_set)
    kmeans.fit(PC_grps)
    kMs_list.append(kmeans.inertia_)
```

```python
print('Minimum inertia', min(kMs_list))
print('Number of clusters :', kMs_list.index(min(kMs_list))+2)
print('_'*50)
print('Maximum inertia', max(kMs_list))
print('Number of clusters :', kMs_list.index(max(kMs_list))+2)
```

```python
ke = KneeLocator(range(2, 10), kMs_list, curve='convex', direction='decreasing')
ke.elbow
```

```python
with plt.style.context('dark_background'):
    plt.plot(range(2, 10), kMs_list, linewidth=3, label='Inertia', alpha=0.7)
    plt.axvline(x=ke.elbow, ls='--', label='elbow', alpha=0.7)
    plt.axvline(x=kMs_list.index(max(kMs_list)) + 2, ls='--', label='maximum of inertia', color='green', alpha=0.7)
    plt.axvline(x=kMs_list.index(min(kMs_list)) + 2, ls='--', label='minimum of inertia', color='red', alpha=0.7)
    plt.axvline(x=3, ls='--', label='i think this is best', color='purple', alpha=0.7)

    plt.xlabel('Number of Clusters')
    plt.ylabel('Inertia')
    plt.title('Inertia vs Number of Clusters', fontweight='bold')
    plt.legend()

    plt.gcf().set_size_inches(10, 4)
    plt.grid(False)

    plt.show()
```

<a id="7.1.2 Silhouette Score for KMeans"></a>
# <left> <p style="background-color:white; font-family:Hack; font-size:70%; color:black; text-align:center; border-radius: 0px; padding:0px;width:30%">7.1.2 Silhouette Score for KMeans</p> </left>

```python
from sklearn.metrics import silhouette_score

kMsSil_score = 0
kMsSil_n = 0
silhouette_coefficient = []

for k in range(2, 10):
    kMs = KMeans(n_clusters=k, **kMs_set).fit(PC_grps)
    score = silhouette_score(PC_grps, kMs.labels_)
    silhouette_coefficient.append(score)

    if score > kMsSil_score:
        kMsSil_score = score
        kMsSil_n = k
    else:
        continue
        
     # Setting dark background style
with plt.style.context('dark_background'):
   
    # Plotting silhouette scores
    plt.figure(figsize=(10, 4))
    plt.plot(range(2, 10), silhouette_coefficient, linewidth=3, marker='o')
    plt.xticks(range(2, 10))
    plt.xlabel('Number Of Clusters')
    plt.ylabel('Silhouette Coefficient')
    plt.title('Silhouette Score for KMeans', fontweight='bold')
    plt.grid(True)
    plt.show()
```

<a id="7.1.3 Calinski-Harabasz Score-KMeans"></a>
# <left> <p style="background-color:white; font-family:Hack; font-size:70%; color:black; text-align:center; border-radius: 0px; padding:0px;width:40%">7.1.3 Calinski-Harabasz Score-KMeans</p> </left>

```python
from sklearn.metrics import calinski_harabasz_score

kMs_calhbz_score = 0
kMs_calhbz_n = 0
calinski_coefficient = []

for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, **kMs_set).fit(PC_grps)
    score = calinski_harabasz_score(PC_grps, kmeans.labels_)
    calinski_coefficient.append(score)

    if score > kMs_calhbz_score:
        kMs_calhbz_score = score
        kMs_calhbz_n = k
    else:
        continue

    # Setting dark background style
with plt.style.context('dark_background'):
    
    # Plotting Calinski-Harabasz scores
    plt.figure(figsize=(10, 4))
    plt.plot(range(2, 10), calinski_coefficient, linewidth=3, marker='o')
    plt.xticks(range(2, 10))
    plt.xlabel('Number Of Clusters')
    plt.ylabel('Calinski-Harabasz Score')
    plt.title('Calinski-Harabasz Score-KMeans', fontweight='bold')
    plt.grid(True)
    plt.show()

print('Calinski-Harabasz Score:', kMs_calhbz_score)
print('Best cluster_n:', kMs_calhbz_n)
```

<a id=" 7.3 DBSCAN"></a>
# <left> <p style="background-color:white; font-family:Hack; font-size:90%; color:black; text-align:center; border-radius: 0px; padding:0px;width:20%"> 7.3 DBSCAN</p> </left>

```python
sil_list = []
calhbz_list = []
DVBDN_list = []
noise_list = []
n_list = []
eps_list = []


for i in range(0, 40, 1):
    e = 0.5 + (i*0.05)
    dbscan = DBSCAN(eps=e)
    dbscan.fit(PC_grps)
    
    labels = dbscan.labels_
    if len(labels) > 1 :
        
        # Number of clusters in labels, ignoring noise if present.
        n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
        n_noise_ = list(labels).count(-1)

        if n_clusters_ > 1 :

            eps_list.append(e)
            n_list.append(n_clusters_)
            noise_list.append(n_noise_)
            sil = silhouette_score(PC_grps, labels)
            davies = davies_bouldin_score(PC_grps, labels)
            calhbz = calinski_harabasz_score(PC_grps, labels)
        
            sil_list.append(sil)
            DVBDN_list.append(davies)
            calhbz_list.append( calhbz)
        
            print("eps = ", e)
            print("Estimated number of clusters: %d" % n_clusters_)
            print("Estimated number of noise points: %d" % n_noise_)
            print('Silhouette : ', sil)
            print('Davis : ', davies)
            print(' calhbz Score : ',    calhbz)
            print('\n____________________________________________________________________\n')
print("END")
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
fig.tight_layout(pad=4.0)
fig.suptitle('DBSCAN', fontsize=35, color='white', fontweight='bold', va='center')


plt.style.use('dark_background')

ax[0, 0].scatter(n_list, sil_list, label='Silhouette_Score', color='blue')  # plot 1: silhouette
ax[0, 0].legend()
ax[0, 0].set_xlabel('Number of Clusters')
ax[0, 0].set_ylabel('Silhouette_Score')
ax[0, 0].set_xticks(range(2, 10, 1))

ax[0, 1].scatter(n_list, DVBDN_list, label='Davies_Score', color='white')  # plot 2: Davies
ax[0, 1].legend()
ax[0, 1].set_xlabel('Number of Clusters')
ax[0, 1].set_ylabel('Davies_Score')
ax[0, 1].set_xticks(range(2, 10, 1))

ax[1, 0].scatter(n_list, calhbz_list, label='Calinski_Harabasz_Score', color='yellow')  # plot 3: Calinski_Harabasz_Score
ax[1, 0].legend()
ax[1, 0].set_xlabel('Number of Clusters')
ax[1, 0].set_ylabel('Calinski_Harabasz_Score')
ax[1, 0].set_xticks(range(2, 10, 1))

ax[1, 1].scatter(n_list, noise_list, label='Noise', color='red')  # plot 4: Noise
ax[1, 1].legend()
ax[1, 1].set_xlabel('Number of Clusters')
ax[1, 1].set_ylabel('Noise')
ax[1, 1].set_xticks(range(2, 10, 1))

plt.show()
```
