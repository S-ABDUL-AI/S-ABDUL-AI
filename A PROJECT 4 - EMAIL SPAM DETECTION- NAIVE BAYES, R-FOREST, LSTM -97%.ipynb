```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.utils import resample

# Assuming 'Spm_Label' is your target variable
# Separate the classes
class_0 = Em_df[Em_df['Spm_Label'] == 0]
class_1 = Em_df[Em_df['Spm_Label'] == 1]

# Check if classes are imbalanced
if len(class_0) != len(class_1):
    # Upsample the minority class (assuming class_1 is the minority class)
    class_1_upsampled = resample(class_1, replace=True, n_samples=len(class_0), random_state=42)

    # Combine the upsampled minority class with the majority class
    balanced_data = pd.concat([class_0, class_1_upsampled])

    # Visualize the class distribution in the balanced data
    fig, ax = plt.subplots(figsize=(8, 4))
    balanced_data['Spm_Label'].value_counts(normalize=True).plot(kind='bar', color=['skyblue', 'navy'])
    plt.title('Email Indicator No(0) and Yes(1) after Oversampling (Balanced Dataset)')
    plt.show()
else:
    print("Classes are already balanced.")
```

# <left> <p style="background-color:darkblue; font-family:Hack; font-size:60%; color:white; text-align:center; border-radius: 0px; padding:10px;width: 30%"> 9.6 Long Shor-Term Memory (LSTM)</p> </left>

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from IPython.display import Markdown, display

# Assuming you have a training dataset X_train and corresponding labels y_train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)

# Tokenize the text
max_words = 10000
tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(X_train)

# Convert text to sequences
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)

# Pad sequences to a fixed length
max_sequence_length = 100  # Choose an appropriate sequence length
X_train_pad = pad_sequences(X_train_seq, maxlen=max_sequence_length)
X_test_pad = pad_sequences(X_test_seq, maxlen=max_sequence_length)

# Build the LSTM model
embedding_dim = 50  # Choose an appropriate embedding dimension
model_lstm = Sequential([
    Embedding(input_dim=max_words, output_dim=embedding_dim, input_length=max_sequence_length),
    LSTM(units=200, dropout=0.5, recurrent_dropout=0.5),
    Dense(units=1, activation='sigmoid')
])

# Compile the model
model_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model_lstm.fit(X_train_pad, y_train, epochs=5, batch_size=64, validation_split=0.1)

# Evaluate the performance of the LSTM model
y_pred_prob = model_lstm.predict(X_test_pad)
y_pred = (y_pred_prob > 0.5).astype(int)

# Training accuracy
y_train_pred = (model_lstm.predict(X_train_pad) > 0.5).astype(int)
train_accuracy = accuracy_score(y_train, y_train_pred)

# Testing accuracy
test_accuracy = accuracy_score(y_test, y_pred)

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Classification report
classification_rep = classification_report(y_test, y_pred)

def print_bold(text):
    display(Markdown(f"**{text}**"))
# Print results
print_bold("Training Accuracy:")
print(train_accuracy)

print_bold("Testing Accuracy:")
print(test_accuracy)

print_bold("Confusion Matrix:")
print(conf_matrix)

print_bold("Classification Report:")
print(classification_rep)
```
