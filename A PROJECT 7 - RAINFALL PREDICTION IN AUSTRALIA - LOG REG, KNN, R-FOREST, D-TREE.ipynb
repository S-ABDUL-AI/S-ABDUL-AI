     ABOUT THE DATASET
</h3>

<blockquote style="background-color: #008080; padding: 10px; border-radius: 5px; color: white;">
    <h3>Data Source</h3>
    <p><strong>Original data source:</strong> Australian Government's Bureau of Meteorology</p>
    <p><strong>Latest data source:</strong> Bureau of Meteorology - Data Weather Observations</p>
    <p><strong>Dataset source:</strong> Rattle - weatherAUS.RData</p>
    <h3>Dataset Description</h3>
    <p>The dataset contains observations of weather metrics for each day from 2008 to 2017. The weatherAUS.csv dataset includes the following fields:</p>
    <ul>
        <li><strong>Date:</strong> Date of the Observation in YYYY-MM-DD (Type: Date)</li>
        <li><strong>Location:</strong> Location of the Observation (Type: Location)</li>
        <li><strong>MinTemp:</strong> Minimum temperature (Unit: Celsius, Type: float)</li>
        <li><strong>MaxTemp:</strong> Maximum temperature (Unit: Celsius, Type: float)</li>
        <li><strong>Rainfall:</strong> Amount of rainfall (Unit: Millimeters, Type: float)</li>
        <li><strong>Evaporation:</strong> Amount of evaporation (Unit: Millimeters, Type: float)</li>
        <li><strong>Sunshine:</strong> Amount of bright sunshine (Unit: hours, Type: float)</li>
        <li><strong>WindGustDir:</strong> Direction of the strongest gust (Type: Compass Points, Type: object)</li>
        <li><strong>WindGustSpeed:</strong> Speed of the strongest gust (Unit: Kilometers/Hour, Type: object)</li>
        <li><strong>WindDir9am:</strong> Wind direction averaged of 10 minutes prior to 9am (Type: Compass Points, Type: object)</li>
        <li><strong>WindDir3pm:</strong> Wind direction averaged of 10 minutes prior to 3pm (Type: Compass Points, Type: object)</li>
        <li><strong>WindSpeed9am:</strong> Wind speed averaged of 10 minutes prior to 9am (Unit: Kilometers/Hour, Type: float)</li>
        <li><strong>WindSpeed3pm:</strong> Wind speed averaged of 10 minutes prior to 3pm (Unit: Kilometers/Hour, Type: float)</li>
        <li><strong>Humidity9am:</strong> Humidity at 9am (Unit: Percent, Type: float)</li>
        <li><strong>Humidity3pm:</strong> Humidity at 3pm (Unit: Percent, Type: float)</li>
        <li><strong>Pressure9am:</strong> Atmospheric pressure reduced to mean sea level at 9am (Unit: Hectopascal, Type: float)</li>
        <li><strong>Pressure3pm:</strong> Atmospheric pressure reduced to mean sea level at 3pm (Unit: Hectopascal, Type: float)</li>
        <li><strong>Cloud9am:</strong> Fraction of the sky obscured by cloud at 9am (Unit: Eights, Type: float)</li>
        <li><strong>Cloud3pm:</strong> Fraction of the sky obscured by cloud at 3pm (Unit: Eights, Type: float)</li>
        <li><strong>Temp9am:</strong> Temperature at 9am (Unit: Celsius, Type: float)</li>
        <li><strong>Temp3pm:</strong> Temperature at 3pm (Unit: Celsius, Type: float)</li>
        <li><strong>RainToday:</strong> If there was rain today (Type: Yes/No, Type: object)</li>
        <li><strong>RISK_MM:</strong> Amount of rain tomorrow (Unit: Millimeters, Type: float)</li>
        <li><strong>RainTomorrow:</strong> If there is rain tomorrow (Type: Yes/No, Type: float)</li>
    </ul>
    <p>Column definitions were gathered from Bureau of Meteorology - Data Weather Observations.</p>

#### 4.3 Computing columns with missing values greater than 5%

```python
# calculating skewness for numerical columns
skewness = RAIN_DATA.select_dtypes(include=['number']).apply(lambda x: x.skew())

# defining a threshold for skewness (e.g., 1 for moderate skewness)
skewness_threshold = 1

# categorizing columns based on skewness and its degree
positively_skewed = skewness[(skewness > skewness_threshold) & (skewness <= 3)]
highly_positively_skewed = skewness[skewness > 3]
negatively_skewed = skewness[(skewness < -skewness_threshold) & (skewness >= -3)]
highly_negatively_skewed = skewness[skewness < -3]
approximately_symmetric = skewness[abs(skewness) <= skewness_threshold]

# creating a DataFrame to store the skewness and degree of skewness information
skewness_info_df = pd.DataFrame({
    "Column Name": skewness.index,
    "Skewness": skewness.values,
    "Skewness Level": ["Highly Positively Skewed" if col in highly_positively_skewed else
                       "Positively Skewed" if col in positively_skewed else
                       "Highly Negatively Skewed" if col in highly_negatively_skewed else
                       "Negatively Skewed" if col in negatively_skewed else
                       "Approximately Symmetric" for col in skewness.index],
    "Degree of Skewness": ["High" if col in highly_positively_skewed or col in highly_negatively_skewed else
                           "Medium" if col in positively_skewed or col in negatively_skewed else
                           "Low" for col in skewness.index]
})

# defining the order of skewness categories
category_order = ["Highly Negatively Skewed", "Negatively Skewed", "Approximately Symmetric", "Positively Skewed", "Highly Positively Skewed"]
degree_of_skewness_order = ["High", "Medium", "Low"]

# sorting the dataFrame by category order and degree of skewness
skewness_info_df["Skewness Level"] = pd.Categorical(skewness_info_df["Skewness Level"], categories=category_order, ordered=True)
skewness_info_df["Degree of Skewness"] = pd.Categorical(skewness_info_df["Degree of Skewness"], categories=degree_of_skewness_order, ordered=True)
skewness_info_df = skewness_info_df.sort_values(by=["Skewness Level", "Degree of Skewness"])

# display the DataFrame
skewness_info_df = skewness_info_df.style.set_properties(**{"background-color": "teal", "color": "white", "border-color": "black"})
skewness_info_df
```

#### 4.5 Checking for outliers in continuous features

```python
# defining a boxplot function
def plot_boxplot(RAIN_DATA, continuous_features):
    # Create a copy of the DataFrame with only the specified continuous_features
    data = RAIN_DATA[continuous_features].copy()
    
    # Creating subplots with the specified number of rows and columns
    fig, axes = plt.subplots(nrows=len(data.columns) // 2, ncols=2, figsize=(10, 15))
    
    # Adjusting the vertical space between subplots
    fig.subplots_adjust(hspace=0.8)
    
    # Defining font properties for titles
    font = {
        'family': 'monospace',
        'color': 'yellow',
        'weight': 'normal',
        'size': 16,
    }
    
    # Generating box plots for each continuous feature
    for ax, feature in zip(axes.flatten(), data.columns):
        sns.boxplot(data[feature], ax=ax)  # Create a box plot for the current feature
        ax.set_title(f' {feature}', fontdict=font)  # Set the title for the box plot
        
    plt.show()  # Displaying the subplots
```

```python
def drop_outliers(RAIN_DATA,outliers): # Defining a function
    
      # Creating a copy of the original dataframe to avoid modifying the original data  
    data = RAIN_DATA.copy()               
    
    # Iterating through each feature specified as an outlier
    for feature in data[outliers].columns: 
        Q1 = data[feature].quantile(0.25) # Calculating the first quartile (Q1) for the feature
        Q3 = data[feature].quantile(0.75) # Calculating the third quartile (Q3) for the feature
        
       # Calculating the Interquartile Range (IQR)  
        IQR = Q3 - Q1                    
        lower_limit = round(Q1 - 1.5 * IQR)  # Calculating the lower limit for identifying outliers
        upper_limit = round(Q3 + 1.5 * IQR) # Calculating the upper limit for identifying outliers
        
        # Replacing values that are greater than the upper limit with the upper limit
        data.loc[data[feature]>= upper_limit,feature] = upper_limit
         
            # Replacing values that are lower than the lower limit with the lower limit
        data.loc[data[feature]<=lower_limit,feature] = lower_limit
        
    return data
```

```python
# Initializing One-hot encoding function for categorical features 
def onehot_encoder(RAIN_DATA, cols):
    data = RAIN_DATA.copy()
    for col in cols:
        # Data in DataFrame and 'col' is the column we will one-hot encode
        dummies = pd.get_dummies(data[col],drop_first=True,prefix=col)
       
        # concatenating dummies and original dataframe
        data = pd.concat([data, dummies], axis=1)
        
        # dropping original columns for which encoding is applied.
        data.drop(col, axis=1,inplace=True)
    return data
```

#### 7.3   Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

# Calculating the confusion matrix
cmtrx = confusion_matrix(y_test, predictions)

# Defining labels for different parts of the confusion matrix
group_names = ['True Neg', 'False Pos', 'False Neg', 'True Pos']

# Calculating the count of each group and format it as a string
group_counts = ["{0:0.0f}".format(value) for value in cmtrx.flatten()]

# Calculating the percentage of each group relative to the total
group_percentages = ["{0:.2%}".format(value) for value in cmtrx.flatten() / np.sum(cmtrx)]

# Combining the group names, counts, and percentages to create labels
labels = [f'{v1}\n{v2}\n{v3}' for v1, v2, v3 in zip(group_names, group_counts, group_percentages)]

# Reshaping the labels to match the shape of the confusion matrix (2x2)
labels = np.asarray(labels).reshape(2, 2)

# Defining categories for the x and y axes
categories = ['No', 'Yes']

# Create a figure size 
plt.figure(figsize=(6, 4))

# Create a heatmap using seaborn
sns.heatmap(cmtrx, annot=labels, fmt='', xticklabels=categories, yticklabels=categories, cmap='Blues')

# Setting labels for x and y
plt.ylabel('True label', fontsize=10, fontweight='bold')
plt.xlabel('Predicted label', fontsize=10, fontweight='bold')

# Setting a plot title 
plt.title('Confusion Matrix', fontsize=20, fontweight='bold')

# Increase the font size and boldness of the category labels on the axes
plt.xticks(fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')

# Show the plot
plt.show()
```
