<a id='2.0'></a>
<h3 style='padding:10px; background:black;color:white;border-radius:0px; text-align:center;'>
    3.0 IMPORTING BRAIN IMAGE DATASETS
</h3>

```python
#setting inital parameters
height, width = 230, 230
batch_size=64
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# defining data directory 
data_dir = 'C:/Users/sherriff/Desktop/PORTFOLIO/COMPUTER VISION/archive/brain_tumor_dataset'

# defining function for image data generation
def image_generator(height, width, batch_size=32):
    datagen = ImageDataGenerator(
        rescale=1./255.,
        validation_split=0.2,
    )
    train_img = datagen.flow_from_directory(
        data_dir,
        batch_size=batch_size,
        subset="training",
        shuffle=True,
        class_mode='binary',
        target_size=(height, width),
        classes={'no': 0., 'yes': 1.}
    )
    validate_img = datagen.flow_from_directory(
        data_dir,
        subset="validation",
        class_mode='binary',
        target_size=(height, width),
        batch_size=batch_size,
        classes={'no': 0., 'yes': 1.}
    )
    return train_img, validate_img

# getting the datasets
train_img, validate_img = image_generator(height, width)

# printing  images
print('\n\n', {'No_brain_tumor_cases': len(np.where(train_img.labels == 0)[0]),
               'brain_tumor_cases': len(np.where(train_img.labels == 1)[0]),
               'No_brain_tumor_cases_val': len(np.where(validate_img.labels == 0)[0]),
               'brain_tumor_cases_val': len(np.where(validate_img.labels == 1)[0])})
```

```python
# defining function to display images 
def display_images_with_renamed_labels(image_generator, num_images=8):
    fig, ax = plt.subplots(2, 4, figsize=(10, 7))
    fig.suptitle("Brain Images")
    
    for k in range(num_images):
        images, labels = image_generator.next()
        i, j = k // 4, k % 4
        
        # renaming labels
        label_name = 'Tumor' if int(labels[0]) == 1 else 'No Tumor'
        
        # setting color based on label
        color = 'red' if int(labels[0]) == 1 else 'green'
        
        ax[i, j].imshow(images[0])
        ax[i, j].set_title(f' {label_name}', color=color)
        ax[i, j].axis('off')
    
    plt.show()

display_images_with_renamed_labels(train_img)
```

<h2 style='padding:10px; background:black;color:white;border-radius:0px; text-align:center;'>
        5.0 MODELING VGG19 TRANSFER LEARNING CONVOLUTIONARY NEURAL NETWORK

<a id=" 5.1 VGG19 Transfer Learning"></a>
# <left> <p style="background-color:black; font-family:Hack; font-size:80%; color:white; text-align:center; border-radius: 0px; padding:10px;width:30%"> 5.1 VGG19 Transfer Learning</p> </left>

```python
# Import necessary libraries
from tensorflow.keras.layers.experimental import preprocessing

# Clear the previous TensorFlow session
tf.keras.backend.clear_session()

# Define input shape
input_shape = (height, width, 3)

# Load VGG19 base model with pre-trained weights
base_model = tf.keras.applications.vgg19.VGG19(
    weights='imagenet',   
    include_top=False,
    input_shape=input_shape
)

# Set base model as non-trainable
base_model.trainable = False

# Create a sequential model
model_vgg19 = tf.keras.Sequential()

# Add VGG19 base model to the sequential model
model_vgg19.add(base_model)

# Flatten the output of the VGG19 base model
model_vgg19.add(tf.keras.layers.Flatten())

# Add a Dense layer with one neuron and sigmoid activation for binary classification
model_vgg19.add(tf.keras.layers.Dense(1, activation='sigmoid'))

# Compile the model with binary crossentropy loss, Adam optimizer, and accuracy metric
model_vgg19.compile(
    loss='binary_crossentropy', 
    optimizer=tf.keras.optimizers.Adam(0.01),
    metrics=['acc']
)

# Display model summary
model_vgg19.summary()
```

<a id=" 5.1.1 VGG19 Model Training with Callbacks"></a>
# <left> <p style="background-color:lightgray; font-family:Hack; font-size:65%; color:black; text-align:center; border-radius: 0px; padding:10px;width:35%"> 5.1.1 VGG19 Model Training with Callbacks</p> </left>

```python
# Defining callbacks for model training
checkpoint = tf.keras.callbacks.ModelCheckpoint(
    'model/vgg19_best.h5',  # Filepath to save the best model weights
    monitor='acc',          # Metric to monitor for saving the best model (accuracy in this case)
    verbose=1,              # Verbosity level (1: update messages)
    mode='max',             # Mode for monitoring ('max' to save the best when the monitored metric increases)
    save_best_only=True      # Save only the best model weights
)

early = tf.keras.callbacks.EarlyStopping(
    monitor="acc",           # Metric to monitor for early stopping (accuracy in this case)
    mode="max",              # Mode for monitoring ('max' to stop training when the monitored metric increases)
    restore_best_weights=True,  # Restore the best weights when early stopping is triggered
    patience=5                # Number of epochs with no improvement after which training will be stopped
)

# Combine callbacks into a list
callbacks_list = [checkpoint, early]

# Training the VGG19 model
history = model_vgg19.fit(
    train_img,                # Training data
    validation_data=validate_img,  # Validation data
    epochs=16,                # Number of training epochs
    shuffle=True,             # Shuffle the training data at the beginning of each epoch
    verbose=True,             # Verbosity level (1: update messages)
    callbacks=callbacks_list  # List of callbacks to apply during training
)
```

<a id="5.1.2 Evaluating the VGG19 Model Results"></a>
# <left> <p style="background-color:lightgray; font-family:Hack; font-size:65%; color:black; text-align:center; border-radius: 0px; padding:10px;width:37%"> 5.1.2 Evaluating the VGG19 Model Results</p> </left>

```python
# evaluating the model on the training and validation image data
train_metrics = model_vgg19.evaluate(train_img)
validate_metrics = model_vgg19.evaluate(validate_img)

# creating a dataFrame for model evaluation  
evaluation_df = pd.DataFrame({'Train Metrics': train_metrics, 'Validate Metrics': validate_metrics}, index=['Loss', 'Accuracy'])

styled_df = evaluation_df.style.set_table_styles(
    [{'selector': 'th',
      'props': [('background-color', 'black'), ('color', 'white')]},
     {'selector': 'td',
      'props': [('background-color', 'black'), ('color', 'white')]}]
)

styled_df
```

<a id="5.1.3 Plotting  Learning Curve of VGG19 Results"></a>
# <left> <p style="background-color:lightgray; font-family:Hack; font-size:65%; color:black; text-align:center; border-radius: 0px; padding:10px;width:40%"> 5.1.3 Plotting  Learning Curve of VGG19 Results</p> </left>

```python
# defining learning_curve(history)
def plot_learning_curve(history):
    # Extracting training and validation accuracy and loss from the history object
    acc = history.history['acc']  
    val_acc = history.history['val_acc']  
    loss = history.history['loss']
    val_loss = history.history['val_loss']
   
    
    # getting the number of epochs
    epochs = range(len(acc))

    # Plotting training and validation accuracy
    plt.plot(epochs, acc, label='Training Accuracy')
    plt.plot(epochs, val_acc, label='Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy-%')
    plt.legend()
    plt.title('Training and Validation Accuracy')
    plt.show()

    # Plotting training and validation loss
    plt.plot(epochs, loss, label='Training Loss')
    plt.plot(epochs, val_loss, label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.title('Training and Validation Loss')
    plt.show()

plot_learning_curve(history)
```

# <left> <p style="background-color:black; font-family:Hack; font-size:80%; color:white; text-align:center; border-radius: 0px; padding:10px;width:55%"> 5.2 Image Augmentation of VGG19 Transfer Learning</p> </left>

<a id=" 5.2.1 Generating Augmented Images for VGG19 Model Training"></a>
# <left> <p style="background-color:lightgray; font-family:Hack; font-size:65%; color:black; text-align:center; border-radius: 0px; padding:10px;width:50%"> 5.2.1 Generating Augmented Images for VGG19 Model Training</p> </left>

```python
# Defining a function to create an augmented image generator
def create_augmented_image_generator(height, width):
    # Configuration for image data augmentation
    datagen = ImageDataGenerator(
        rescale=1./255.,
        width_shift_range=0.1,
        height_shift_range=0.1,
        shear_range=0.1,
        zoom_range=0.1,
        rotation_range=30,
        horizontal_flip=True,
        brightness_range=(0.5, 1.0)
    )
    
    # Creating an augmented training data generator using flow_from_directory
    train_img_aug = datagen.flow_from_directory(
        data_dir,
        batch_size=64,
        shuffle=True,
        class_mode='binary',
        target_size=(height, width),
        classes={'no': 0., 'yes': 1.}
    )
    
    return train_img_aug

# Creating an instance of the augmented training data generator
train_img_aug = create_augmented_image_generator(height, width)
```

<a id=" 5.2.2 Augmented Image VGG19 Model Training with Callbacks"></a>
# <left> <p style="background-color:lightgray; font-family:Hack; font-size:65%; color:black; text-align:center; border-radius: 0px; padding:10px;width:50%"> 5.2.2 Augmented Image VGG19 Model Training with Callbacks</p> </left>

```python
history = model_vgg19.fit(
    train_img_aug,
    validation_data=validate_img,
    epochs=3,
    verbose=True,
    callbacks=callbacks_list,
    batch_size=32  # Adjusting the batch size
)
```

<a id=" 6.1 InceptionV3 Model Initialization"></a>
# <left> <p style="background-color:black; font-family:Hack; font-size:80%; color:white; text-align:center; border-radius: 0px; padding:10px;width:37%"> 6.1 InceptionV3 Model Initialization</p> </left>

<a id=" 6.1.2 InceptionV3 Image Generation"></a>
# <left> <p style="background-color:lightgray; font-family:Hack; font-size:65%; color:black; text-align:center; border-radius: 0px; padding:10px;width:33%"> 6.1.2 InceptionV3 Image Generation</p> </left>

```python
# InceptionV3 Model Initialization

# setting up model and image data generator 
height = 230
width = 230
train_img, validate_img = image_generator(height, width)

# deleting  previous TensorFlow session
tf.keras.backend.clear_session()

# setting up input shape and load InceptionV3 base model
input_shape = (height, width, 3)
base_model = tf.keras.applications.InceptionV3(
    weights='imagenet', 
    include_top=False, 
    input_shape=input_shape
)
base_model.trainable = False

# building the InceptionV3 model
model_inceptionv3 = tf.keras.Sequential()
model_inceptionv3.add(base_model)
model_inceptionv3.add(tf.keras.layers.Flatten())
model_inceptionv3.add(tf.keras.layers.Dense(1, activation='sigmoid'))

# compiling the model
model_inceptionv3.compile(
    loss='binary_crossentropy', 
    optimizer=tf.keras.optimizers.Adam(0.001),
    metrics=['acc']
)

# checking model summary
model_inceptionv3.summary()
```

<a id=" 6.1.3 InceptionV3 Model Training with Callbacks"></a>
# <left> <p style="background-color:lightgray; font-family:Hack; font-size:65%; color:black; text-align:center; border-radius: 0px; padding:10px;width:39%"> 6.1.3 InceptionV3 Model Training with Callbacks</p> </left>

```python
# training  the InceptionV3 model 
checkpoint = tf.keras.callbacks.ModelCheckpoint('model/inceptionv3_best.h5', monitor='acc', verbose=1, mode='max',save_best_only=True)
early = tf.keras.callbacks.EarlyStopping(monitor="acc", mode="max",restore_best_weights=True, patience=5)
callbacks_list = [checkpoint,early]

history = model_inceptionv3.fit(
        train_img,
        validation_data=validate_img,
        epochs=16, 
        verbose=True,
        callbacks=callbacks_list)
```

<a id="6.1.4 Evaluating the InceptionV3 Model Results"></a>
# <left> <p style="background-color:lightgray; font-family:Hack; font-size:65%; color:black; text-align:center; border-radius: 0px; padding:10px;width:39%"> 6.1.4 Evaluating the InceptionV3 Model Results</p> </left>

```python
# evaluating the model on the training and validation sets
train_metrics = model_inceptionv3.evaluate(train_img)
validate_metrics = model_inceptionv3.evaluate(validate_img)

# creating a dataframe for evaluation metrics
evaluation_df = pd.DataFrame({'Train Metrics': train_metrics, 'Validate Metrics': validate_metrics}, index=['Loss', 'Acc'])

# customizing the dataFrame
styled_df = evaluation_df.style.set_table_styles(
    [{'selector': 'th',
      'props': [('background-color', 'black'), ('color', 'white')]},
     {'selector': 'td',
      'props': [('background-color', 'black'), ('color', 'white')]}]
)

styled_df
```

```python
# creating an instance of image augmentation generator
train_img_aug = create_augmented_image_generator(height, width)

# defining Callbacks
early = tf.keras.callbacks.EarlyStopping(monitor="val_loss", mode="min", restore_best_weights=True, patience=5)

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor='acc', factor=0.3, patience=3, min_lr=0.0000001)

# combining callbacks 
callbacks_list = [checkpoint, early, reduce_lr]

# training the image augmented model 
history = model_inceptionv3.fit(
    train_img_aug,
    validation_data=validate_img,
    epochs= 16,
    verbose=True,
    callbacks=callbacks_list,
    batch_size=32  # adjusted the batch size
)
```
