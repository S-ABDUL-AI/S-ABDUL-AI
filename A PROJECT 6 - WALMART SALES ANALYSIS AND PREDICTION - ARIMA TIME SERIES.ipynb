<a id=" 7.5 Investigating Stationarity"></a>
# <left> <p style="background-color:white; font-family:Hack; font-size:70%; color:brown; text-align:center; border-radius: 0px; padding:0px;width:30%"> 7.5 Investigating Stationarity</p> </left>

```python
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller

data = WLmT_week['LogWeekly_Sales']

# Performing the ADF test
result = adfuller(data)

# Extracting and print the test statistic and p-value
test_statistic = result[0]
p_value = result[1]

# Define ANSI escape codes for colors
darkblue = '\033[34m'  
reset_color = '\033[34m'  

# Print the results in navy blue
print(f'{darkblue}Test Statistic: {test_statistic}{reset_color}')
print(f'{darkblue}P-Value: {p_value}{reset_color}')

# Interpreting the results
if p_value <= 0.05:
    print(f'{darkblue}The time series is stationary (reject the null hypothesis).{reset_color}')
else:
    print(f'{darkblue}The time series is non-stationary (fail to reject the null hypothesis).{reset_color}')
```

<a id=" 7.6 Strenghtening Stationarity"></a>
# <left> <p style="background-color:white; font-family:Hack; font-size:70%; color:brown; text-align:center; border-radius: 0px; padding:0px;width:30%"> 7.6 Strenghtening Stationarity</p> </left>

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller

# Original time series
original_series = WLmT_week['LogWeekly_Sales']

# Step 1: Differencing
diff = original_series.diff().dropna()

# Step 2: Log Transformation
diff_log = np.log(diff)

# Handle missing or non-finite values
diff_log = diff_log.dropna()  # Remove missing values
diff_log = diff_log.replace([np.inf, -np.inf], np.nan).dropna()  # Remove non-finite values

# Visualization
plt.figure(figsize=(12, 6))
plt.subplot(411)
plt.plot(original_series, label="Original Series", color="orange")  # Set color to orange
plt.legend()

plt.subplot(412)
plt.plot(diff, label="Differenced Series", color="orange")  # Set color to orange
plt.legend()

plt.subplot(413)
plt.plot(diff_log, label="Differenced and Log-Transformed Series", color="orange")  # Set color to orange
plt.legend()

plt.show()
```

<h3 style='padding:8px; background:#0071ce;color:white;border-radius:0px; text-align:center;'>
    8.0 TIME SERIES MODELING
</h3>

<a id=" 8.1 Train Test Split"></a>
# <left> <p style="background-color:white; font-family:Hack; font-size:70%; color:brown; text-align:center; border-radius: 0px; padding:0px;width:20%"> 8.1 Train Test Split</p> </left>

```python
# Defining the split point (e.g., 80% for training, 20% for testing)
split_point = int(0.8 * len(WLmT_week))
```

```python
# Splitting the data into training and testing sets
train_data = WLmT_week[:split_point]
test_data = WLmT_week[split_point:]
```

```python
# Performing differencing to make the series stationary
d = 1
train_diff = train_data.diff(periods=d).dropna()
test_diff = test_data.diff(periods=d).dropna()
```

<a id=" 8.2 Auto-ARIMA MODEL"></a>
# <left> <p style="background-color:white; font-family:Hack; font-size:70%; color:brown; text-align:center; border-radius: 0px; padding:0px;width:25%"> 8.2 Auto-ARIMA MODEL</p> </left>

```python
# Using the auto_arima function to automatically select the best ARIMA model
model_auto_arima = auto_arima(train_diff, trace=True, start_p=0, start_q=0, start_P=0, start_Q=0,
              max_p=20, max_q=20, max_P=20, max_Q=20, seasonal=True, maxiter=200,
              information_criterion='aic', stepwise=False, suppress_warnings=True, D=1, max_D=10,
              error_action='ignore', approximation=False)
model_auto_arima.fit(train_diff)
```

<a id='9.0'></a>
<h3 style='padding:8px; background:#0071ce;color:white;border-radius:0px; text-align:center;'>

<a id=" 9.1 Forecasting with Auto-ARIMA"></a>
# <left> <p style="background-color:white; font-family:Hack; font-size:70%; color:brown; text-align:center; border-radius: 0px; padding:0px;width:30%"> 9.1 Forecasting with Auto-ARIMA</p> </left>

```python
y_pred = model_auto_arima.predict(n_periods=len(test_diff))
y_pred = pd.DataFrame(y_pred, index=test_diff.index, columns=['Prediction'])
```

```python
# Defining colors for the train and test lines
train_color = 'blue'
test_color = 'green'

# Defining the background colors
figure_background_color = 'black'  # Background color of the entire figure
plot_background_color = 'white'  # Background color of the plot area

fig, ax = plt.subplots(figsize=(20, 6))
fig.set_facecolor(figure_background_color)  # Set the background color of the entire figure

ax.set_title('Prediction of Weekly Sales Using Auto-ARIMA', fontsize=25, fontweight= 'bold')

# Plotting the train data with the specified color
plt.plot(train_diff, label='Train', color=train_color)

# Plotting the test data with the specified color
plt.plot(test_diff, label='Test', color=test_color)

# Plotting the prediction of ARIMA
plt.plot(y_pred, label='Prediction of ARIMA', color= 'orange')

ax.set_facecolor(plot_background_color)  # Set the background color of the plot area

plt.legend(loc='best', fontsize=20)
plt.xlabel('Date', fontsize=16, fontweight= 'bold')
plt.ylabel('Weekly Sales', fontsize=16, fontweight= 'bold')
plt.xticks(rotation=0, fontsize=12, fontweight='bold')
plt.yticks(rotation=0, fontsize=12, fontweight='bold')
plt.show()
```

<a id=" 9.2 Forecasting with ExponentialSmoothing"></a>
# <left> <p style="background-color:white; font-family:Hack; font-size:70%; color:brown; text-align:center; border-radius: 0px; padding:0px;width:40%"> 9.2 Forecasting with ExponentialSmoothing</p> </left>

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Defining  colors for the train and test lines
train_color = 'blue'
test_color = 'green'

# Defining the background colors
figure_background_color = 'black'  # Background color of the entire figure
plot_background_color = 'white'  # Background color of the plot area

fig, ax = plt.subplots(figsize=(20, 6))
fig.set_facecolor(figure_background_color)  # Set the background color of the entire figure

ax.set_title('Prediction of Weekly Sales using Exponential Smoothing', fontsize=25, fontweight= 'bold')

# Plotting the train data with the specified color
plt.plot(train_diff, label='Train', color=train_color)

# Plotting the test data with the specified color
plt.plot(test_diff, label='Test', color=test_color)

# Plotting the prediction using Exponential Smoothing
plt.plot(y_pred, label='Prediction using Exponential Smoothing', color = 'orange')
ax.set_facecolor(plot_background_color)  # Set the background color of the plot area

plt.legend(loc='best', fontsize=20)
plt.xlabel('Date', fontsize=16, fontweight= 'bold')
plt.ylabel('Weekly Sales', fontsize=16, fontweight= 'bold')
plt.xticks(rotation=0, fontsize=12, fontweight='bold')
plt.yticks(rotation=0, fontsize=12, fontweight='bold')
plt.show()
```

<a id=" 9.3 Model Evaluation with MSE, MAE and RMSE"></a>
# <left> <p style="background-color:white; font-family:Hack; font-size:70%; color:brown; text-align:center; border-radius: 0px; padding:0px;width:40%"> 9.3 Model Evaluation with MSE, MAE and RMSE</p> </left>

```python
# Removing rows with missing values from both y_pred and test_diff
y_pred = y_pred.dropna()
test_diff = test_diff.dropna()

# Aligning the lengths of test_diff and y_pred
min_length = min(len(test_diff), len(y_pred))
test_diff = test_diff[:min_length]
y_pred = y_pred[:min_length]

# Calculating evaluation metrics
mse = mean_squared_error(test_diff, y_pred)
mae = mean_absolute_error(test_diff, y_pred)
rmse = np.sqrt(mse)
```

```python
print(f"Mean Squared Error (MSE): {mse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```
